## Macierze

### Operacje na obiektach

1. Skalowanie
    $$
    f(
        \begin{bmatrix}
        x \\
        y
        \end{bmatrix}
    )
    =
    \begin{bmatrix}
        s_x * x \\
        s_y * y
    \end{bmatrix}
    $$
2. Rotacja
    $$
    f(
        \begin{bmatrix}
        x \\
        y
        \end{bmatrix}
    )
    =
    \begin{bmatrix}
        x * cos(\theta) - y * sin(\theta) \\
        x * sin(\theta) + y * cos(\theta)
    \end{bmatrix}
    $$
3. Translacja
    $$
    f(
        \begin{bmatrix}
        x \\
        y
        \end{bmatrix}
    )
    =
    \begin{bmatrix}
        x + 1 \\
        y - 1
    \end{bmatrix}
    $$


### Macierze operacji 2D

#### Translacji

$$
T = 
\begin{bmatrix}
    1 & 2 \\
    3 & 4
\end{bmatrix}
\text{lub dla } 
\vec{V}=(T_x, T_y),
T=
\begin{bmatrix} 1 & 0 & T_x \\ 0 & 1 & T_y \\ 0 & 0 & 1 \end{bmatrix}
$$

### Skalowania

$$
S = \begin{bmatrix}
        s_x & 0 \\
        0 & s_y
    \end{bmatrix}
\text{lub }
\begin{bmatrix} s_x & 0 & 0 \\ 0 & s_y & 0 \\ 0 & 0 & 1 \end{bmatrix}
$$

### Rotacji

$$
R = \begin{bmatrix}
        cos(\theta) & -sin(\theta) \\
        sin(\theta) & cos(\theta)
    \end{bmatrix}
\text{lub }
\begin{bmatrix} cos(\theta) & -sin(\theta) & 0 \\ sin(\theta) & cos(\theta) & 0 \\ 0 & 0 & 1 \end{bmatrix}
$$


### Konkatenacja operacji

Mając dwie matryce operacji można je połączyć w jedną.

Cykl operacji:

1. $A \begin{bmatrix}x \\ y \end{bmatrix} = \begin{bmatrix}x_1 \\ y_1 \end{bmatrix}$
2. $B \begin{bmatrix}x_1 \\ y_1 \end{bmatrix} = \begin{bmatrix}x_2 \\ y_2 \end{bmatrix}$

Można skrócić jako:

$$
B A \begin{bmatrix}x \\ y \end{bmatrix}
$$


### Macierze operacji 3D

#### Rotacji

$$
R_x(\theta) = 
\begin{bmatrix}
    1 & 0 & 0 \\
    0 & cos(\theta) & -sin(\theta) \\
    0 & sin(\theta) & cos(\theta) 
\end{bmatrix}
$$

$$
R_y(\theta) = 
\begin{bmatrix}
    cos(\theta) & 0 & sin(\theta) \\
    0 & 1 & 0 \\
    -sin(\theta) & 0 & cos(\theta) 
\end{bmatrix}
$$

$$
R_z(\theta) = 
\begin{bmatrix}
    cos(\theta) & -sin(\theta) & 0 \\
    sin(\theta) & cos(\theta) & 0 \\
    0 & 0 & 1
\end{bmatrix}
$$

### Translacji

$$
\text{Dla wektora }
\vec{V}=(T_x, T_y, T_z),
T=
\begin{bmatrix} 1 & 0 & 0, & T_x \\ 0 & 1 & 0 & T_y \\ 0 & 0 & 1 & T_z \\ 0 & 0 & 0 & 1 \end{bmatrix}
$$

### Skalowania

$$
S = \begin{bmatrix} s_x & 0 & 0 & 0 \\ 0 & s_y & 0 & 0 \\ 0 & 0 & s_z & 0 \\ 0 & 0 & 0 & 1 \end{bmatrix}
$$

## Pipeline

Dla widoku wielkości $n_x * n_y$
$$
M_{vp} = \begin{bmatrix} 
    \frac{n_x}{2} & 0 & 0 & \frac{n_x - 1}{2} \\
    0 & \frac{n_y}{2} & 0 & \frac{n_y - 1}{2} \\
    0 & 0 & 1 & 0 \\
    0 & 0 & 0 & 1 
\end{bmatrix}
$$

Dla przestrzeni widoku ortogonalnego stworzonego przez trzy punkty:

- $(l,b,n)$ w lewym dolnym rogu
- $(r,b,n)$ w prawym dolnym rogu
- $(r,t,f)$ w prawym górnym rogu

$$
M_{ort} = \begin{bmatrix} 
    \frac{2}{r - l} & 0 & 0 & -\frac{l + r}{r - l} \\
    0 & \frac{2}{t - b} & 0 & -\frac{b + t}{t - b} \\
    0 & 0 & \frac{2}{n - f} & -\frac{n + f}{n - f} \\
    0 & 0 & 0 & 1 
\end{bmatrix}
$$

$$
M_{per} = M_{ort} \cdot \begin{bmatrix} 
    n & 0 & 0 & 0 \\
    0 & n & 0 & 0 \\
    0 & 0 & n + f & -nf \\
    0 & 0 & 0 & 1 
\end{bmatrix}
$$

Dla wektorów określających kierunek patrzenia kamery: $\vec{g}$ $\vec{v}$ $\vec{u}$ oraz $\vec{e}$ określającym przesunięcie kamery

$$
M_{cam} = \begin{bmatrix} 
    u_x & u_y & u_z & 0 \\
    v_x & v_y & v_z & 0 \\
    w_x & w_y & w_z & 0 \\
    0 & 0 & 0 & 1 
\end{bmatrix}
\begin{bmatrix}
    1 & 0 & 0 & -e_x \\
    0 & 1 & 0 & -e_y \\
    0 & 0 & 1 & -e_z \\
    0 & 0 & 0 & 1 
\end{bmatrix}
$$

**Rendering pipeline**:

$$
pipeline(\begin{bmatrix}x \\ y \\ z\\ 1\end{bmatrix}) = M_{vp}M_{ort}M_{cam}\begin{bmatrix}x \\ y \\ z\\ 1\end{bmatrix}
$$
